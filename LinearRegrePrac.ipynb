```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel("Area(sqr ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area,df.price,color='red',marker='+')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_ # m-gradient
```

```python
reg.intercept_ # y interrcept
```

```python
reg.predict([[5000]])
```

```python
plt.xlabel("Area(sqr ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='b')
plt.show()
```

```python
df=pd.read_csv('homeprices_new.csv')
df
```

```python
import math
median_room=math.floor(df.bedrooms.median())
median_room
```

```python
df.bedrooms=df.bedrooms.fillna(median_room)
```

```python
reg.coef_
```

```python
reg.intercept_
```
