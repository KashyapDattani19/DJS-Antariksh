```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

Bar Graph

```python
x=["kd","rg","pd","hk"]
y=[100,95,80,85]

c=['g','b','pink','y']

plt.xlabel("Student initials",fontsize=15)
plt.ylabel("Marks in Python",fontsize=15)
plt.title("Student Data",fontsize=30)

plt.bar(x,y,color=c,width=0.5,align='edge',edgecolor='r',linewidth=5,linestyle=':',alpha=0.5) # or align='center'
plt.show()                                                                                    # alpha is responsible for the darkness for the bar

```

```python
x=['First','Second','Third','Fouth']
y=[80,65,90,70]
z=[30,15,45,40]

p=np.arange(len(x))
width=0.2
p1=[j+width for j in p]

plt.xlabel("Year Quater",fontsize=15)
plt.ylabel("Sales percentage",fontsize=15)
plt.title("Company Data",fontsize=30)

plt.bar(p,y,width,color="blue",label="Company1")
plt.bar(p1,z,width,color="green",label="Company2")

plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```

```python
x=['First','Second','Third','Fouth']
y=[80,65,90,70]
z=[30,15,45,40]

p=np.arange(len(x))
width=0.2
p1=[j+width for j in p]

plt.xlabel("Year Quater",fontsize=15)
plt.ylabel("Sales percentage",fontsize=15)
plt.title("Company Data",fontsize=30)

plt.barh(x,y,width,color="blue",label="Company1")
plt.barh(x,z,width,color="green",label="Company2")


plt.legend()
plt.show()
```

Scatter Graph

```python
day=[1,2,3,4,5,6,7]
num=[2,5,1,7,8,5,9]
num2=[1,2,3,4,5,6,7]

colors=['red','blue','green','yellow','pink','blue','orange']
sizes=[70,50,20,80,300,100,50]

plt.xlabel("Day")
plt.ylabel("Number")
plt.title("Scatter Graph", fontsize=20)

plt.scatter(day,num,c=colors,s=sizes,alpha=0.7,marker="*")
plt.scatter(day,num2,s=sizes,alpha=1)
plt.show()

#Also has colormap attribute
```

PieChart

```python
x=[10,20,30,40]
y=["c","c++","java","python"]
ex=[0.3,0.0,0.0,0.0]
c=['r','b','g','y']

plt.title("Languages")

plt.pie(x,labels=y,explode=ex,colors=c,
        autopct="%1.1f%%",shadow=True,radius=1,
        labeldistance=1.1,startangle=90,
        textprops={"fontsize":12},counterclock=False,wedgeprops={"linewidth":3,"edgecolor":"magenta"},
        rotatelabels=False)
plt.legend(loc=3)
plt.show()
```

```python
plt.pie([1])  #Dot piechart
plt.show()
```

```python
#Donut piechart

x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]


plt.pie(x,labels=y,radius=1.5)
# plt.pie(x1,radius=1,colors=c)
plt.pie([1],colors="w")  #Dot piechart

plt.show()
```

```python
x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]


plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```

Stem Plot

```python
x=[1,2,3,4,5,6]
y=[2,2,5,7,3,8]

plt.stem(x,y,linefmt=":",markerfmt="rx",bottom=4,basefmt="g",label="Python") #orientation="horizontal"
plt.legend()
plt.show()
```

Box plot

```python
x=[10,20,30,40,50,60,70,120]
y=[10,20,30,40,50,60,70,90]
z=[x,y]
plt.boxplot(z,labels=["Python","C++"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="blue"))          #notch=True,vert=False,widths=0.4,patch_artist=True,whis=2,flierprops=dict(markeredgecolor="b")
      

plt.show()
```

Stack plot

```python
x=[10,20,30,40,50]
y=[10,20,40,20,50]
y1=[10,20,40,20,50]
y2=[20,30,50,30,60]
l=["python","c","java"]

plt.stackplot(x,y,y1,y2,labels=l,baseline="zero")        #colors=["r","g","y"]
plt.legend()
plt.show()
```

Step Plot

```python
x=[1,2,3,4,5,6]
y=[2,2,3,5,4,6]

plt.title("Python")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.step(x,y,color="r",marker="o",ms=10,mfc="b",label="new")
# plt.grid()
plt.legend()
plt.show()
```

Fill Between

```python
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,color="b")

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4),alpha=0.5)

plt.show()
```
