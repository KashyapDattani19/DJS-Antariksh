```python
var1={"Emp ID":["E01","E02","E03","E04","E05"],
     "Name":["Kashyap","Agrim","Shubham","Harsh","Krish"],
     "Age":[20,30,40,25,50]}

var2={"Emp ID":["E01","E02","E03","E04","E05"],
     "Salary":[90000,70000,65000,75000,80000]
}

df1=pd.DataFrame(var1)
df2=pd.DataFrame(var2)

print(df1)
print(df2)
```

```python
print(pd.merge(df1,df2))
```

```python
var3={"Emp ID":["E01","E02","E03","E04","E05"],
     "Name":["Kashyap","Agrim","Shubham","Harsh","Krish"],
     "Age":[20,30,40,25,50]}

var4={"Emp ID":["E06","E07","E08","E09","E10"],
     "Name":["Kashyap","Agrim","Shubham","Harsh","Krish"],
     "Age":[20,30,40,25,50]}

df3=pd.DataFrame(var3)
df4=pd.DataFrame(var4)

print(pd.concat([df3,df4]))
```
